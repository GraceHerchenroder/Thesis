```python
# Importing libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt
```

```python
# Performing two-way ANOVA
def statAnova_ut(kinematic_data):
  dataframe = pd.DataFrame({'Gender': np.repeat(['women', 'men'], 5),
                          'Size': ['small', 'small', 'small', 'large', 'large', 'small', 'small', 'small', 'large', 'large'],
                          'Kinematic': kinematic_data[0][:]})
  group_1_mean = np.mean(kinematic_data[0][:3])
  group_1_sd = np.std(kinematic_data[0][:3])
  group_2_mean = np.mean(kinematic_data[0][3:5])
  group_2_sd = np.std(kinematic_data[0][3:5])
  group_3_mean = np.mean(kinematic_data[0][5:8])
  group_3_sd = np.std(kinematic_data[0][5:8])
  group_4_mean = np.mean(kinematic_data[0][8:])
  group_4_sd = np.std(kinematic_data[0][8:])
  print(kinematic_data[0][:3])
  print(group_1_mean, group_1_sd)
  print(kinematic_data[0][3:5])
  print(group_2_mean, group_2_sd)
  print(kinematic_data[0][5:8])
  print(group_3_mean, group_3_sd)
  print(kinematic_data[0][8:])
  print(group_4_mean, group_4_sd)
  print(dataframe)
  model = ols('Kinematic ~ C(Gender) + C(Size) + C(Gender):C(Size)', data=dataframe).fit()
  result = sm.stats.anova_lm(model, typ=2)
  return result

def statAnova_lt(kinematic_data):
  dataframe = pd.DataFrame({'Gender': np.repeat(['women', 'men'], 5),
                          'Size': ['small', 'small', 'small', 'large', 'large', 'small', 'small', 'small', 'large', 'large'],
                          'Kinematic': kinematic_data[1][:]})
  group_1_mean = np.mean(kinematic_data[1][:3])
  group_1_sd = np.std(kinematic_data[1][:3])
  group_2_mean = np.mean(kinematic_data[1][3:5])
  group_2_sd = np.std(kinematic_data[1][3:5])
  group_3_mean = np.mean(kinematic_data[1][5:8])
  group_3_sd = np.std(kinematic_data[1][5:8])
  group_4_mean = np.mean(kinematic_data[1][8:])
  group_4_sd = np.std(kinematic_data[1][8:])
  print(kinematic_data[1][:3])
  print(group_1_mean, group_1_sd)
  print(kinematic_data[1][3:5])
  print(group_2_mean, group_2_sd)
  print(kinematic_data[1][5:8])
  print(group_3_mean, group_3_sd)
  print(kinematic_data[1][8:])
  print(group_4_mean, group_4_sd)
  print(dataframe)
  model = ols('Kinematic ~ C(Gender) + C(Size) + C(Gender):C(Size)', data=dataframe).fit()
  result = sm.stats.anova_lm(model, typ=2)
  return result

def statAnova_lum(kinematic_data):
  dataframe = pd.DataFrame({'Gender': np.repeat(['women', 'men'], 5),
                          'Size': ['small', 'small', 'small', 'large', 'large', 'small', 'small', 'small', 'large', 'large'],
                          'Kinematic': kinematic_data[2][:]})
  group_1_mean = np.mean(kinematic_data[2][:3])
  group_1_sd = np.std(kinematic_data[2][:3])
  group_2_mean = np.mean(kinematic_data[2][3:5])
  group_2_sd = np.std(kinematic_data[2][3:5])
  group_3_mean = np.mean(kinematic_data[2][5:8])
  group_3_sd = np.std(kinematic_data[2][5:8])
  group_4_mean = np.mean(kinematic_data[2][8:])
  group_4_sd = np.std(kinematic_data[2][8:])
  print(kinematic_data[2][:3])
  print(group_1_mean, group_1_sd)
  print(kinematic_data[2][3:5])
  print(group_2_mean, group_2_sd)
  print(kinematic_data[2][5:8])
  print(group_3_mean, group_3_sd)
  print(kinematic_data[2][8:])
  print(group_4_mean, group_4_sd)
  print(dataframe)
  model = ols('Kinematic ~ C(Gender) + C(Size) + C(Gender):C(Size)', data=dataframe).fit()
  result = sm.stats.anova_lm(model, typ=2)
  return result

def box_whisker(kinematic_data):
   small_men_data_ut = kinematic_data[0][8:]
   large_men_data_ut = kinematic_data[0][5:8]
   small_women_data_ut = kinematic_data[0][:3]
   large_women_data_ut = kinematic_data[0][3:5]
   small_men_data_lt = kinematic_data[1][8:]
   large_men_data_lt = kinematic_data[1][5:8]
   small_women_data_lt = kinematic_data[1][:3]
   large_women_data_lt = kinematic_data[1][3:5]
   small_men_data_lum = kinematic_data[2][8:]
   large_men_data_lum = kinematic_data[2][5:8]
   small_women_data_lum = kinematic_data[2][:3]
   large_women_data_lum = kinematic_data[2][3:5]
   fig, ax = plt.subplots(3, 2, figsize=(6, 10))
   ax[0][0].boxplot([small_men_data_ut, large_men_data_ut], labels = ['small', 'large'])
   ax[0][0].set_title('Men')
   ax[0][0].set_ylabel('Upper Thoracic')
   ax[0][1].boxplot([small_women_data_ut, large_women_data_ut], labels = ['small', 'large'])
   ax[0][1].set_title('Women')
   ax[1][0].boxplot([small_men_data_lt, large_men_data_lt], labels = ['small', 'large'])
   ax[1][0].set_title('Men')
   ax[1][0].set_ylabel('Lower Thoracic')
   ax[1][1].boxplot([small_women_data_lt, large_women_data_lt], labels = ['small', 'large'])
   ax[1][1].set_title('Women')
   ax[2][0].boxplot([small_men_data_lum, large_men_data_lum], labels = ['small', 'large'])
   ax[2][0].set_title('Men')
   ax[2][0].set_ylabel('Lumbar')
   ax[2][1].boxplot([small_women_data_lum, large_women_data_lum], labels = ['small', 'large'])
   ax[2][1].set_title('Women')
   plt.show()
```

```python
# Follow Up analysis: Significant
def significant_ut(kinematic_data):
  small_men_data = kinematic_data[0][8:]
  large_men_data = kinematic_data[0][5:8]
  small_women_data = kinematic_data[0][:3]
  large_women_data = kinematic_data[0][3:5]

  [stat_small, p_small] = stats.ttest_ind(small_men_data, small_women_data);
  [stat_large, p_large] = stats.ttest_ind(large_men_data, large_women_data);
  [stat_women, p_women] = stats.ttest_ind(small_women_data, large_women_data);
  [stat_men, p_men] = stats.ttest_ind(small_men_data, large_men_data);

  result = multipletests([p_small, p_large, p_women, p_men], alpha = 0.05, method = 'bonferroni')

  return result[1]

def significant_lt(kinematic_data):
  small_men_data = kinematic_data[1][8:]
  large_men_data = kinematic_data[1][5:8]
  small_women_data = kinematic_data[1][:3]
  large_women_data = kinematic_data[1][3:5]

  [stat_small, p_small] = stats.ttest_ind(small_men_data, small_women_data);
  [stat_large, p_large] = stats.ttest_ind(large_men_data, large_women_data);
  [stat_women, p_women] = stats.ttest_ind(small_women_data, large_women_data);
  [stat_men, p_men] = stats.ttest_ind(small_men_data, large_men_data);

  result = multipletests([p_small, p_large, p_women, p_men], alpha = 0.05, method = 'bonferroni')

  return result[1]

def significant_lum(kinematic_data):
  small_men_data = kinematic_data[2][8:]
  large_men_data = kinematic_data[2][5:8]
  small_women_data = kinematic_data[2][:3]
  large_women_data = kinematic_data[2][3:5]

  [stat_small, p_small] = stats.ttest_ind(small_men_data, small_women_data);
  [stat_large, p_large] = stats.ttest_ind(large_men_data, large_women_data);
  [stat_women, p_women] = stats.ttest_ind(small_women_data, large_women_data);
  [stat_men, p_men] = stats.ttest_ind(small_men_data, large_men_data);

  result = multipletests([p_small, p_large, p_women, p_men], alpha = 0.05, method = 'bonferroni')

  return result[1]
```
